# Dynamics simulations

Running adaptive sampling can be seen equal as running production runs in SchNarc, however, during adaptive sampling the mean of more models is used for propagation of nuclei, which is not the case here, but can be done if wished. 

This section will not contain examples in this notebook. To find out how to do dynamics with learned NACs or NACs approximated from Hessians, please go into the folder "LearnedNacs" or "withHessianApprox", respectively. Go into each trajectory folder and execute the "run.sh" file.
