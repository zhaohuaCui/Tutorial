# Training, Validation, and Prediction

In this notebook you will find out how to train, validate, and test a SchNarc model. The examples will be done using only few data points to make the training feasible during this tutorial.
In this folder, you can also find scripts that show the executions you have to carry out in order to train, validate, and test your models.

## Training

We will start with training a SchNarc model on the previously generated data set of 100 data points. We will use the min-loss for this purpose. In SchNarc, there are several arguments available that you can set, such as "--min_loss" to use the phase-free training algorithm.

```python
import os
```

Please terminate the training after you can find a 
#### "best_model"
in your "Train" folder (usually 1-2 minutes).

```python
#Model path: This is the name of the path where the "best_model" is saved
modelpath = "./Train"

# point to an existing data base
DB = "../DBs/Fulvene.db"

# specify your SCHNARC path
SCHNARC = "../../src/scripts/"


#Train your model and use schnet as a descriptor
# --split specifies the number of training and validation points (the rest is used for testing)
# you can set --cuda if you have a GPU
# you also need a tradeoffs.yaml file in which you specify which properties to train

os.system("python %s/run_schnarc.py train schnet %s %s --split 80 10 --min_loss --tradeoffs tradeoffs.yaml --batch_size 2 " %(SCHNARC,DB,modelpath))
```

##  Validation

You could see above how a model can be trained. Now we will predict the error on the validation set.
Therefore, we change the mode from "train" to "eval" and the "split" from numbers to "validation" ("test", "training", or "all" can be set). 

The model is saved in the Train-folder and named "best_model".


```python
os.system("python %s/run_schnarc.py eval schnet %s %s --split validation --batch_size 2 " %(SCHNARC,DB,modelpath))
```

The predictions are saved in the Train-folder and are named "evaluation_values.npz". The reference data is saved as "evaluation_qmvalues.npz".
The (not phase-free) errors are saved in the file "evaluation.csv".


### Analyse validation

```python
# Load evaluation
import numpy as np
import matplotlib.pyplot as plt

path = "Train/"
qmeval = np.load("%s/evaluation_qmvalues_.npz"%path,allow_pickle=True)
eval = np.load("%s/evaluation_values_.npz"%path,allow_pickle=True)

# Plot energies eval
energies_qme = qmeval['energy'].reshape(-1,2)
energies_e = eval['energy'].reshape(-1,2)
for state in range(2):
    plt.scatter(energies_qme[:,state],energies_e[:,state],label="S$_%i$"%state)
plt.legend()
plt.xlabel("Energy Data point QM [H]")
plt.ylabel("Energy Data point [H]")
plt.show()
```

Above you can see a scatter plot. Note that we only trained on very few data and for a short time, so we do not expect better results at this stage.

## Prediction

We can now predict a new data base and predict the properties of the molecules in this data base.
We will generate structures with Wigner sampling implemented in SHARC. Therefore, specify your SHARC path below.

```python
# first we generate many initial conditions
SHARC="/user/julia/software/SchNarc/sharc/source/../bin/"

os.system("python2 %s/wigner.py -n 1000 -x ../Inputfiles/freq.molden"%SHARC)

#this function writes the structures into "initconds.xyz".

# then we transform the initconds.xyz file to an ase-db format
# Note that we need to transform the structures from Angstrom to Bohr

from ase.units import Bohr
import ase.io
from ase.atoms import Atoms

geoms = ase.io.read("initconds.xyz",":")
geoms_bohr = []
for i in range(len(geoms)):
    geoms_bohr.append(Atoms(geoms[i].get_atomic_numbers(),geoms[i].get_positions()/Bohr))

# write db
ase.io.write("../DBs/initconds.db", geoms_bohr)

```

We can now predict the properties of the newly generated data set. Note that we will predict the whole training set and will not need to specify a split.

```python
os.system("python %s/run_schnarc.py pred ../DBs/initconds.db %s --batch_size 2 " %(SCHNARC,modelpath))
```

The predictions can be found in the file "predictions.npz". Below you will see an example on how to analyse the predictions.
